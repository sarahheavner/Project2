```python
# Dependencies
import pandas as pd
```

```python
# Read csv file with the pandas library
aqi_df = pd.read_csv("New_Data/aqi2019.csv")

aqi_df.head()
```

```python
#remove unnecessary columns
clean_aqi = aqi_df[['us_state', 'county', 'days_with_aqi', 'good_days', 'moderate_days', 'unhealthy_sensitive_days', 'unhealthy_days', 'very_unhealthy_days', 'hazardous_days']]
clean_aqi.head()
```

```python
# Read csv file with the pandas library
poverty_df = pd.read_csv("New_Data/poverty2019.csv")

poverty_df.head()
```

```python
#merge dataframes on state and county
combined_df = pd.merge(clean_aqi, poverty_df, how="left", on=['us_state', 'county'])
combined_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
combined_df.to_csv("Data/combineddata.csv", index=False, header=True)
```
